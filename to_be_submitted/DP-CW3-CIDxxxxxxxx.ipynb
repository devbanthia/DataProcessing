# 9.2 Proximal Operators Related to ℓ1-norm

```julia
z = A.Q9_2_a_data_z
gamma = A.Q9_2_a_data_gamma
lambda = A.Q9_2_a_data_lambda

c = gamma * lambda
x = similar(z)           

for i in eachindex(z)
    if z[i] > c
        x[i] = z[i] - c
    elseif z[i] < -c
        x[i] = z[i] + c
    else
        x[i] = 0
    end
end

A.Q9_2_a_ans_x = x



z = A.Q9_2_b_data_z
a = A.Q9_2_b_data_a
gamma = A.Q9_2_b_data_gamma
lambda = A.Q9_2_b_data_lambda

a_matrix = Diagonal(a)
c = lambda * gamma 

z_projection = inv(a_matrix) * z
x = similar(z)           
threshold = c * diag(inv(a_matrix) .^ 2)

for i in eachindex(z_projection)
    if z_projection[i] > threshold[i]
        x[i] = z_projection[i] - threshold[i]

    elseif z_projection[i] < -1 * threshold[i]
        x[i] = z_projection[i] + threshold[i]
    else
        x[i] = 0
    end
end

A.Q9_2_b_ans_x = x




z = A.Q9_2_c_data_z
U = A.Q9_2_c_data_U
gamma = A.Q9_2_c_data_gamma
lambda = A.Q9_2_c_data_lambda

z_projection = U' * z

c = gamma*lambda
x = similar(z_projection) 

for i in eachindex(z_projection)
    if z_projection[i] > c
        x[i] = z_projection[i] - c
    elseif z_projection[i] < -c
        x[i] = z_projection[i] + c
    else
        x[i] = 0
    end
end

A.Q9_2_c_ans_x = x


z = A.Q9_2_d_data_z
U = A.Q9_2_d_data_U
gamma = A.Q9_2_d_data_gamma
lambda = A.Q9_2_d_data_lambda

c = gamma * lambda
projection = U * z
soft_thresholded = sign.(projection) .* max.(abs.(projection) .- c, 0.0)
print(soft_thresholded)

result = U' * soft_thresholded

A.Q9_2_d_ans_x = result

```

# 10.1 Sparsity under Wavelet Transform

```julia

using FFTW
using Wavelets, WaveletsExt

function wavet(X)
    return dwt(X, wavelet(WT.haar))
end

function iwavet(Z)
    return idwt(Z, wavelet(WT.haar))
end

function opA(X, Omega)
    O1 = fft(X)
    O2 = fftshift(O1)
    O3 = Omega .* O2
    y = []
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[j, i] != 0)
                push!(y, O3[j, i])
                cnt = cnt + 1
            end
        end
    end

    return y
end

function opA_hat(Y, Omega)
    O0 = iwavet(Y)
    O1 = fft(O0)
    O2 = fftshift(O1)
    O3 = Omega .* O2
    y = []
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[j, i] != 0)
                push!(y, O3[j, i])
                cnt = cnt + 1
            end
        end
    end

    return y
end



function opAadj_hat(y, Omega)
 
    O1_init = zeros(ComplexF64, size(Omega)[1], size(Omega)[2])
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[j, i] != 0)
                O1_init[j, i] = y[cnt]
                cnt = cnt + 1
            end
        end
    end
    O1 = O1_init
    O2 = ifftshift(O1)
    O3 = ifft(O2)
    O4 = wavet(O3)

    return O4
end

function opAadj(y, Omega)
    O1_init = zeros(ComplexF64, size(Omega)[1], size(Omega)[2])
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[j, i] != 0)
                O1_init[j, i] = y[cnt]
                cnt = cnt + 1
            end
        end
    end
    O1 = O1_init
    O2 = ifftshift(O1)
    O3 = ifft(O2)
    return O3
end
```

```julia
using LinearAlgebra

X0 = A.Q10_1_data_X0
Omega = A.Q10_1_data_Omega
y = A.Q10_1_data_y
lambda = A.Q10_1_data_lambda
lambda = 20

function stc(w, threshold)
    magnitudes = abs.(w)
    soft_magnitudes = max.(magnitudes .- threshold, 0)
    result = soft_magnitudes .* exp.(im * angle.(w))
    return result
end



tau = 1
Y0 = wavet(X0)
Y_k = copy(Y0)
X_k = copy(X0)
println(size(wavet(X_k)))

q_X = (-1 * opAadj_hat((y - opA_hat(Y_k, Omega)), Omega))

A.Q10_1_1_ans_gradient = q_X
A.Q10_1_2_ans_L = 1

X_k_prev = iwavet(copy(X_k))
Z =  Y_k - tau * (-1 * opAadj_hat((y - opA_hat(Y_k, Omega)), Omega))
Y_k = stc.(Z, tau * lambda)
X_k = iwavet(Y_k)

A.Q10_1_3_ans_X1 = X_k

g = Float64[]  

for i in 1:10
    X_k_prev = iwavet(copy(X_k))
    Z =  Y_k - tau * (-1 * opAadj_hat((y - opA_hat(Y_k, Omega)), Omega))
    Y_k = stc.(Z, tau * lambda)
    X_k = iwavet(Y_k)
    
    grad_h_Y = lambda * sign.(Y_k)
    g_k = iwavet(lambda * sign.(Y_k))
    push!(g, norm(g_k))
    
end

A.Q10_1_4_ans_g = g


Y0 = wavet(X0)
Y_k = copy(Y0)
X_k = copy(X0)
lambda = 0.1
for i in 1:100000
    
    X_k_prev = iwavet(copy(Y_k))
    Z =  Y_k - tau * (-1 * opAadj_hat((y - opA_hat(Y_k, Omega)), Omega))
    Y_k = stc.(Z, tau * lambda)
    X_k = iwavet(Y_k)

    g_k = iwavet(lambda * sign.(Y_k))

    print("$i: ")
    print(lambda * norm(X_k - X_k_prev))
    print("         ")
    print(norm(X_k)/1000)
    println("")
    if((lambda * norm(X_k - X_k_prev)) < (norm(X_k)/1000))
        break
    end
end

```

# 10.1 Sparsity under Wavelet Transform

```julia
# Pkg.add("FFTW")
using FFTW
using Wavelets, WaveletsExt

# Q.6.1.1 Compute y
function wavet(X)
    # Using WT.haar yields same result as WT.db1
    return dwt(X, wavelet(WT.haar))
end

function iwavet(Z)
    return idwt(Z, wavelet(WT.haar))
end

function opA(X, Omega)
    # Apply 2D FFT
    O1 = fft(X)
    # Apply 2D FFT Shift
    O2 = fftshift(O1)
    # Apply element-wise mask
    O3 = Omega .* O2
    # Vectorise O3
    y = []
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[i, j] != 0)
                push!(y, O3[i, j])
                cnt = cnt + 1
            end
        end
    end

    # return O1, O2, y
    return y
end

function opA_hat(Y, Omega)
    # Apply 2D FFT
    O0 = iwavet(Y)
    O1 = fft(O0)
    # Apply 2D FFT Shift
    O2 = fftshift(O1)
    # Apply element-wise mask
    O3 = Omega .* O2
    # Vectorise O3
    y = []
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[i, j] != 0)
                push!(y, O3[i, j])
                cnt = cnt + 1
            end
        end
    end

    # return O1, O2, y
    return y
end


# O1, O2, y = opA(data.Q6_1_data_X, data.Q6_1_data_Omega)

function opAadj_hat(y, Omega)
    # Reconstruct y into a sparse matrix
    O1_init = zeros(ComplexF64, size(Omega)[1], size(Omega)[2])
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[i, j] != 0)
                O1_init[i, j] = y[cnt]
                cnt = cnt + 1
            end
        end
    end
    O1 = O1_init
    # fftshift is self-adjoint
    O2 = ifftshift(O1)
    # Adjoint of FFT is its inverse
    O3 = ifft(O2)
    O4 = wavet(O3)

    # return O1, O2, O3

    return O4
end

function opAadj(y, Omega)
    # Reconstruct y into a sparse matrix
    O1_init = zeros(ComplexF64, size(Omega)[1], size(Omega)[2])
    cnt = 1
    for i in 1:size(Omega)[1]
        for j in 1:size(Omega)[2]
            if(Omega[i, j] != 0)
                O1_init[i, j] = y[cnt]
                cnt = cnt + 1
            end
        end
    end
    O1 = O1_init
    # fftshift is self-adjoint
    O2 = ifftshift(O1)
    # Adjoint of FFT is its inverse
    O3 = ifft(O2)

    # return O1, O2, O3

    return O3
end
```

```julia
using FFTW
using Wavelets
using LinearAlgebra


X0 = data.Q10_1_data_X0
omega = data.Q10_1_data_Omega
y = data.Q10_1_data_y
lambda = data.Q10_1_data_lambda


function dct_matrix(n::Int)
    D = zeros(Float64, n, n) 
    for k in 1:n
        for j in 1:n
            if k == 1
                D[k, j] = sqrt(1 / n) 
            else
                D[k, j] = sqrt(2 / n) * cos((π * (j - 0.5) * (k - 1)) / n)
            end
        end
    end
    return D
end

function haar_wavelet_matrix(n::Int)

    function haar_wavelet_recursive(size::Int)
        if size == 1
            return [1.0]
        else
            H = haar_wavelet_recursive(size ÷ 2)
            top = kron(H, [1.0 1.0]) / sqrt(2.0)  
            bottom = kron(Matrix{Float64}(I, size ÷ 2, size ÷ 2), [1.0 -1.0]) / sqrt(2.0)
            return vcat(top, bottom)
        end
    end
    
    return haar_wavelet_recursive(n)
end

function dft_matrix(n::Int)
    ω = exp(-2im * π / n)  
    scale_factor = 1 / sqrt(n)  
    return [scale_factor * ω^(j * k) for j in 0:n-1, k in 0:n-1]
end

function rearrange(y::Vector, omega, order::Int64)

    result = zeros(eltype(y), size(omega))  
    indices = findall(omega .== 1)  
    indices = order == 0 ? indices : sort(indices, by = x -> (x[2], x[1]))

    result[indices] .= y

    return result
end


function soft_threshold(w, threshold)

    magnitudes = abs.(w)
    soft_magnitudes = max.(magnitudes .- threshold, 0)
    result = soft_magnitudes .* exp.(im * angle.(w))
    
    return result

end

function normalize_columns!(A::Matrix{T}) where T<:Number
    for j in 1:size(A, 2)
        col = @view A[:, j]
        norm = sqrt(sum(abs2, col))


        if norm > 0
            col ./= norm
        end
    end
    return A
end


wt = haar_wavelet_matrix(128)
dft = dft_matrix(128)
sparse_DFT_matrix = omega .* fftshift(dft)
y = rearrange(y, omega, 0)


grad_X = sparse_DFT_matrix * (y - sparse_DFT_matrix*X0)

L = 1
tau = 1/L

X = X0 - tau * grad_X

wavelet_transform = wt*X

wavelets_thresholded = soft_threshold(wavelet_transform, lambda)
X = wt' * wavelets_thresholded 


L = 1
tau = 1/L

X = X0

for i in 1:10

    println(norm(y - sparse_DFT_matrix*X))
    grad_X = sparse_DFT_matrix' * (y - sparse_DFT_matrix*X)

    X = X + tau * grad_X
    wavelet_transform = wt * X
    wavelets_thresholded = soft_threshold(wavelet_transform, lambda)
    X = wt' * wavelets_thresholded 


end
println(X[1, :])

using Images
using FFTW
using Plots
println(size(y))
p1 = plot(heatmap(abs.(ifft(y)), color=:grays), title="Original Image", legend=false)
p2 = plot(heatmap(abs.(ifft(X)), color=:grays), title="Reconstructed Image", legend=false)

plot(p1, p2)

```
